##### depicts class imbalance in the data.
##### Thus, we use undersampling as a technique to handle imbalanced data

```python
yes=data[data.Bankrupt==1]
no=data[data.Bankrupt==0]
```

```python
no_sample=no.sample(220)
```

```python
print(no_sample.shape)
```

```python
bankrupt_data=pd.concat([yes,no_sample],axis=0)
bankrupt_data.head(5)
```

```python
bankrupt_data.tail(5)
```

```python
from sklearn.model_selection import train_test_split
x=bankrupt_data.drop(['Company_id', 'Bankrupt'],axis=1)
y=bankrupt_data.Bankrupt
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
feature_name=x.columns
```

There are 94 features explaining the target variable (Bankruptcy). Thus,LDA (Linear Discriminant Analysis) is a dimensionality reduction technique  is used to transform a high-dimensional dataset into a lower-dimensional space.The focusis on to retain top 15% of features explaning the target variable.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
lda=LDA()
x_train_lda=lda.fit_transform(x_train,y_train)
x_test_lda=lda.transform(x_test)
```

```python
lda_coefficients = np.abs(lda.coef_).mean(axis=0)

# Create a DataFrame to display feature importance
coef_df = pd.DataFrame({
    'Feature': feature_name,
    'Importance': lda_coefficients
})

# Sort features by importance
coef_df = coef_df.sort_values(by='Importance', ascending=False)

# Select the top 30% of features
top_30_percent = int(len(feature_name) * 0.3)
top_features = coef_df.head(top_30_percent)

print("Top 30% features based on LDA importance:")
print(top_features)
```

```python
#### Reduce x_train and x_test to the top features
selected_features = top_features['Feature'].values
x_train_selected = pd.DataFrame(x_train, columns=feature_name)[selected_features].values
x_test_selected = pd.DataFrame(x_test, columns=feature_name)[selected_features].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

reg=LogisticRegression()
reg.fit(x_train_selected,y_train)
reg_score=accuracy_score(y_test,reg.predict(x_test_selected))
reg_score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,reg.predict(x_test_selected))
print(report)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train_selected,y_train)
tree_score=accuracy_score(y_test,tree.predict(x_test_selected))
tree_score
```

```python
report=classification_report(y_test,tree.predict(x_test_selected))
print(report)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=50)
rf.fit(x_train_selected,y_train)
rf_score=accuracy_score(y_test,rf.predict(x_test_selected))
rf_score
```

```python
report=classification_report(y_test,rf.predict(x_test_selected))
print(report)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(n_estimators=100)
xgb.fit(x_train_selected,y_train)
xgb_score=accuracy_score(y_test,xgb.predict(x_test_selected))
xgb_score
```

```python
report=classification_report(y_test,xgb.predict(x_test_selected))
print(report)
```

```python
model_comparison=pd.DataFrame({'Model':['Logistic_Regression','Decision_Tree','Random_Forest','XGB'],
                                'Accuracy_score':[reg_score, tree_score, rf_score, xgb_score]})
model_comparison.sort_values('Accuracy_score', ascending=0)
```

By comparing the performance of all the ML models, we can come to a conclusion that the best models which gives the highest accuracy score and a descent precision and recall are Random Forest and XGB classifier with an accuracy score of 85.2273%.
